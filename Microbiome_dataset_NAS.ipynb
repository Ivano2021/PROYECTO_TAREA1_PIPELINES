```r
# =========================
# High NAS metagenomics
# Exploratory analysis in R
# =========================

# Instalar paquetes si hace falta
# install.packages(c("readxl", "dplyr", "tidyr", "ggplot2", "pheatmap", "stringr", "forcats", "scales"))

library(readxl)
library(dplyr)
library(tidyr)
library(ggplot2)
library(pheatmap)
library(stringr)
library(forcats)
library(scales)

# -------------------------
# 1. Leer archivo
# -------------------------
file_path <- "/content/HighNAS_papernetwork_data.xlsx"

# Ver nombres de hojas
excel_sheets(file_path)

# Cambiá el nombre de la hoja si hace falta
df <- read_excel(file_path, sheet = 1, col_names = TRUE)

# Ver estructura
str(df)
head(df)

# -------------------------
# 2. Renombrar primera columna
# -------------------------
colnames(df)[1] <- "Feature"

# Convertir todo excepto Feature a numérico
df <- df %>%
  mutate(across(-Feature, as.numeric))

# -------------------------
# 3. Separar fila técnica de burden
# -------------------------
variant_row <- df %>%
  filter(str_detect(Feature, "^Variants_per_million_reads$"))

species_df <- df %>%
  filter(!str_detect(Feature, "^Variants_per_million_reads$")) %>%
  # Nos quedamos solo con especies y excluimos IPR/COG
  filter(!str_detect(Feature, "^IPR"),
         !str_detect(Feature, "^COG"))

# -------------------------
# 4. Pasar a formato largo
# -------------------------
species_long <- species_df %>%
  pivot_longer(
    cols = -Feature,
    names_to = "Sample",
    values_to = "Count"
  )

# -------------------------
# 5. Total de lecturas por muestra
# -------------------------
total_by_sample <- species_long %>%
  group_by(Sample) %>%
  summarise(Total_reads = sum(Count, na.rm = TRUE)) %>%
  arrange(desc(Total_reads))

print(total_by_sample)

# Gráfico 1: carga microbiana total por muestra
g_total <- ggplot(total_by_sample, aes(x = reorder(Sample, -Total_reads), y = Total_reads)) +
  geom_col() +
  theme_bw() +
  labs(
    title = "Total de lecturas microbianas por muestra (High NAS)",
    x = "Muestra",
    y = "Lecturas totales"
  )

print(g_total)

ggsave("HighNAS_total_reads_per_sample.png", g_total, width = 8, height = 5, dpi = 300)

# -------------------------
# 6. Top especies globales
# -------------------------
top_species <- species_long %>%
  group_by(Feature) %>%
  summarise(Total_count = sum(Count, na.rm = TRUE)) %>%
  arrange(desc(Total_count))

print(top_species)

top10_species <- top_species %>%
  slice_head(n = 10)

# Gráfico 2: top 10 especies más abundantes
g_top10 <- ggplot(top10_species, aes(x = fct_reorder(Feature, Total_count), y = Total_count)) +
  geom_col() +
  coord_flip() +
  theme_bw() +
  labs(
    title = "Top 10 especies más abundantes (High NAS)",
    x = "Especie",
    y = "Lecturas totales"
  )

print(g_top10)

ggsave("HighNAS_top10_species.png", g_top10, width = 8, height = 6, dpi = 300)

# -------------------------
# 7. Abundancia relativa por muestra
# -------------------------
species_rel <- species_long %>%
  left_join(total_by_sample, by = "Sample") %>%
  mutate(RelAbundance = ifelse(Total_reads > 0, Count / Total_reads, 0))

# Top 10 por abundancia global para stacked barplot
top10_names <- top10_species$Feature

species_rel_top10 <- species_rel %>%
  mutate(Feature_plot = ifelse(Feature %in% top10_names, Feature, "Otros")) %>%
  group_by(Sample, Feature_plot) %>%
  summarise(RelAbundance = sum(RelAbundance), .groups = "drop")

# Gráfico 3: composición relativa por muestra
g_stack <- ggplot(species_rel_top10, aes(x = Sample, y = RelAbundance, fill = Feature_plot)) +
  geom_col() +
  theme_bw() +
  scale_y_continuous(labels = percent_format()) +
  labs(
    title = "Composición relativa de especies por muestra (High NAS)",
    x = "Muestra",
    y = "Abundancia relativa",
    fill = "Taxón"
  )

print(g_stack)

ggsave("HighNAS_relative_abundance_top10.png", g_stack, width = 10, height = 6, dpi = 300)

# -------------------------
# 8. Prevalencia (% de muestras donde aparece cada especie)
# -------------------------
prevalence_df <- species_long %>%
  group_by(Feature) %>%
  summarise(
    Samples_present = sum(Count > 0, na.rm = TRUE),
    Prevalence = Samples_present / n_distinct(species_long$Sample)
  ) %>%
  arrange(desc(Prevalence), desc(Samples_present))

print(prevalence_df)

top_prev <- prevalence_df %>%
  slice_head(n = 15)

# Gráfico 4: prevalencia
g_prev <- ggplot(top_prev, aes(x = fct_reorder(Feature, Prevalence), y = Prevalence)) +
  geom_col() +
  coord_flip() +
  theme_bw() +
  scale_y_continuous(labels = percent_format()) +
  labs(
    title = "Prevalencia de especies en muestras High NAS",
    x = "Especie",
    y = "% de muestras con presencia"
  )

print(g_prev)

ggsave("HighNAS_prevalence_top15.png", g_prev, width = 8, height = 6, dpi = 300)

# -------------------------
# 9. Core microbiome exploratorio
# Definición: presentes en >= 50% de muestras
# -------------------------
core_species <- prevalence_df %>%
  filter(Prevalence >= 0.5)

print(core_species)

write.csv(core_species, "HighNAS_core_species.csv", row.names = FALSE)

# -------------------------
# 10. Heatmap de especies más abundantes
# -------------------------
heatmap_species <- top_species %>%
  slice_head(n = 20) %>%
  pull(Feature)

heatmap_mat <- species_df %>%
  filter(Feature %in% heatmap_species) %>%
  column_to_rownames("Feature") %>%
  as.matrix()

# Transformación log para estabilizar
heatmap_mat_log <- log10(heatmap_mat + 1)

pheatmap(
  heatmap_mat_log,
  scale = "row",
  cluster_rows = TRUE,
  cluster_cols = TRUE,
  fontsize_row = 8,
  main = "Heatmap de las 20 especies más abundantes (High NAS)",
  filename = "HighNAS_heatmap_top20_species.png",
  width = 8,
  height = 8
)

# -------------------------
# 11. Relación exploratoria con mutational burden
# -------------------------
if (nrow(variant_row) == 1) {
  burden_long <- variant_row %>%
    pivot_longer(
      cols = -Feature,
      names_to = "Sample",
      values_to = "Variants_per_million_reads"
    )

  # Carga microbiana total vs burden
  burden_vs_microbes <- total_by_sample %>%
    left_join(burden_long, by = "Sample")

  print(burden_vs_microbes)

  cor_test_total <- cor.test(
    burden_vs_microbes$Total_reads,
    burden_vs_microbes$Variants_per_million_reads,
    method = "spearman"
  )

  print(cor_test_total)

  g_corr <- ggplot(
    burden_vs_microbes,
    aes(x = Total_reads, y = Variants_per_million_reads, label = Sample)
  ) +
    geom_point(size = 3) +
    geom_text(nudge_y = 2, size = 3) +
    theme_bw() +
    labs(
      title = "Relación exploratoria entre carga microbiana y burden somático",
      x = "Lecturas microbianas totales",
      y = "Variants per million reads"
    )

  print(g_corr)

  ggsave("HighNAS_total_microbes_vs_burden.png", g_corr, width = 7, height = 5, dpi = 300)

  # Correlación con especies específicas relevantes
  species_of_interest <- c(
    "Escherichia__coli",
    "Enterococcus__faecalis",
    "Cutibacterium__acnes",
    "Malassezia__restricta",
    "Salmonella__enterica"
  )

  species_interest_long <- species_long %>%
    filter(Feature %in% species_of_interest) %>%
    left_join(burden_long, by = "Sample")

  # Tabla de correlaciones por especie
  cor_species <- species_interest_long %>%
    group_by(Feature) %>%
    summarise(
      rho = suppressWarnings(cor(Count, Variants_per_million_reads, method = "spearman")),
      p_value = suppressWarnings(cor.test(Count, Variants_per_million_reads, method = "spearman")$p.value),
      .groups = "drop"
    ) %>%
    arrange(p_value)

  print(cor_species)
  write.csv(cor_species, "HighNAS_species_vs_burden_spearman.csv", row.names = FALSE)
}

# -------------------------
# 12. Exportar tablas útiles
# -------------------------
write.csv(total_by_sample, "HighNAS_total_reads_per_sample.csv", row.names = FALSE)
write.csv(top_species, "HighNAS_top_species_global.csv", row.names = FALSE)
write.csv(prevalence_df, "HighNAS_species_prevalence.csv", row.names = FALSE)

# -------------------------
# 13. Resumen rápido en consola
# -------------------------
cat("\n===== RESUMEN RÁPIDO =====\n")
cat("Número de muestras:", ncol(species_df) - 1, "\n")
cat("Número de especies:", nrow(species_df), "\n")
cat("Top 5 especies globales:\n")
print(top_species %>% slice_head(n = 5))
cat("\nCore microbiome exploratorio (>=50% de muestras):\n")
print(core_species)
```

```r
#DIVERSIDAD MICRBIANA (ALFA Y BETA)
# Paquetes
packages <- c("readxl", "dplyr", "vegan", "ggplot2", "tibble")
for (p in packages) {
  if (!require(p, character.only = TRUE)) {
    install.packages(p, repos = "https://cloud.r-project.org/")
    library(p, character.only = TRUE)
  }
}

# Leer Excel correctamente
df <- read_excel("/content/HighNAS_papernetwork_data.xlsx", sheet = 1)

# Renombrar primera columna
colnames(df)[1] <- "Feature"

# Pasar columnas numéricas
df <- df %>%
  mutate(across(-Feature, as.numeric))

# Quedarse solo con especies
species_df <- df %>%
  filter(!grepl("^Variants_per_million_reads$", Feature),
         !grepl("^IPR", Feature),
         !grepl("^COG", Feature))

# Matriz especie x muestra
mat <- species_df %>%
  tibble::column_to_rownames("Feature") %>%
  as.matrix()

# Transponer: filas = muestras, columnas = especies
mat_t <- t(mat)

# -------------------------
# Diversidad alfa
# -------------------------

# Shannon
shannon <- vegan::diversity(mat_t, index = "shannon")

# Chao1
chao1 <- vegan::estimateR(mat_t)[2, ]

alpha_df <- data.frame(
  Sample = rownames(mat_t),
  Shannon = shannon,
  Chao1 = chao1
)

print(alpha_df)

# Gráfico Shannon
g_shannon <- ggplot(alpha_df, aes(x = Sample, y = Shannon)) +
  geom_col() +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 90, hjust = 1)) +
  ggtitle("Diversidad alfa (Shannon)")

print(g_shannon)

# Gráfico Chao1
g_chao1 <- ggplot(alpha_df, aes(x = Sample, y = Chao1)) +
  geom_col() +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 90, hjust = 1)) +
  ggtitle("Riqueza estimada (Chao1)")

print(g_chao1)

# -------------------------
# Diversidad beta
# -------------------------

# Bray-Curtis
dist_bc <- vegan::vegdist(mat_t, method = "bray")

# PCoA
pcoa <- cmdscale(dist_bc, k = 2)

pcoa_df <- data.frame(
  Sample = rownames(mat_t),
  PC1 = pcoa[,1],
  PC2 = pcoa[,2]
)

g_pcoa <- ggplot(pcoa_df, aes(x = PC1, y = PC2, label = Sample)) +
  geom_point(size = 3) +
  geom_text(vjust = -0.5) +
  theme_minimal() +
  ggtitle("PCoA - Bray-Curtis")

print(g_pcoa)
```
